```python

# Create feature selection pipeline  (Categorical Features - Ordinal Encoding)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1 = Pipeline(steps=[('imputer',SimpleImputer(strategy='mean')),
                           ('Scaler',StandardScaler())])
```

```python
cat_pipe1 = Pipeline(steps=[('imputer',SimpleImputer(strategy='most_frequent')),
                            ('ordinal',OrdinalEncoder())])
```

```python
pre1 = ColumnTransformer([('num',num_pipe1,con),
                          ('cat',cat_pipe1,cat)]).set_output(transform='pandas')
```

```python
pre1
```

```python
X_pre = pre1.fit_transform(X)
X_pre.head()
```

```python
# Feature selection 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
lr = LinearRegression()
sel = SequentialFeatureSelector(lr,direction = 'forward')
sel.fit_transform(X_pre,Y)
sel_cols =sel.get_feature_names_out()
sel_cols
```

```python
len(sel_cols)
```

```python
sel_cols[0].split('__')[1]
```

```python
imp_cols = []
for i in sel_cols:
    s = i.split('__')[1]
    imp_cols.append(s)

imp_cols
    
```

```python
# Get above in Dataframe 

X_sel = X[imp_cols]
X_sel
```
